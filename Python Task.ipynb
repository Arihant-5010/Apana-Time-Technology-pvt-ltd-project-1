```python
import pandas as pd
data=pd.read_excel("orginal data.xlsx")
data
```

```python
data.head
```

```python
data.isnull().sum()
```

```python
# Fill NaN with mean for all numerical columns
clean_data=data.fillna(data.mean(numeric_only=True))
```

```python
clean_data
```

```python
clean_data.isnull().sum()
```

```python
import numpy as np
numeric_data = clean_data.select_dtypes(include=[np.number])
numeric_data
```

```python
mean_values = numeric_data.mean()
mean_values
```

```python
median_values = numeric_data.median()
median_values
```

```python
mode_values = numeric_data.mode().iloc[0]
mode_values
```

```python
variance_values = numeric_data.var()
variance_values
```

```python
std_dev=numeric_data.std()
std_dev
```

```python
range_values = numeric_data.max() - numeric_data.min()
range_values
```

```python
# Skewness and Kurtosis
from scipy.stats import skew
skewness_values = numeric_data.apply(skew)
skewness_values

```

```python
from scipy.stats import kurtosis
kurtosis_values = numeric_data.apply(kurtosis)
kurtosis_values
```

```python
# Combine all results into a DataFrame for better readability
import pandas as pd
eda_results = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values,
    'Variance': variance_values,
    'Std Dev': std_dev,
    'Range': range_values,
    'Skewness': skewness_values,
    'Kurtosis': kurtosis_values
})

# Display the EDA results
print(eda_results)
```

```python
# Plot Skewness and Kurtosis
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
eda_results[['Skewness', 'Kurtosis']].plot(kind='bar', figsize=(14,6), title="Skewness and Kurtosis")
plt.ylabel("Values")
plt.xlabel("Columns")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```
